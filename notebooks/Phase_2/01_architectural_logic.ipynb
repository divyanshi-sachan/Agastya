# ARCHITECTURE LOGIC 

##  1. Core Idea
While transformer-based models such as BERT provide strong contextual representations, they are inherently semantic in nature and do not explicitly encode structural properties of text. Legal clauses, however, exhibit both semantic and structural patterns, where features such as clause length and syntactic complexity often correlate with clause type.

To address this limitation, we propose a hybrid architecture that augments transformer embeddings with structured auxiliary signals through a dedicated fusion mechanism. This introduces an inductive bias that enables the model to jointly reason over semantic content and structural characteristics.

##  2. Custom Architectural Block 

Rather than naïvely concatenating the clause-length feature with the transformer output, we design a **feature-aware gating mechanism** that dynamically modulates the influence of structural information.

Let $h_{BERT} \in \mathbb{R}^d$ denote the contextual embedding from Legal-BERT, and $l \in \mathbb{R}$ denote the normalized clause length.

We compute a gating coefficient:
$$g = \sigma(W_l \cdot l + b)$$

The final representation is defined as:
$$h_{final} = h_{BERT} \odot g + h_{BERT}$$
where $\odot$ denotes element-wise scaling.

This mechanism allows the model to learn when structural features should influence the decision, rather than treating them as uniformly important across all inputs.

##  3. Why this is IMPORTANT 
This design introduces a structural inductive bias into the model. While standard transformers rely purely on data-driven learning, our approach guides the model to consider clause-level structural properties that are known to be relevant in legal reasoning.

**For example:**
- Longer clauses often correspond to liability or detailed obligations.
- Short clauses often correspond to metadata (dates, names).

By explicitly modeling this relationship, the architecture improves generalization in cases where semantic signals alone are insufficient.

##  4. Optimization-Level Innovation
In addition to architectural modifications, we further optimize the learning process by incorporating **class-weighted cross-entropy loss** to address imbalance in the CUAD dataset.

This modification aligns the training objective with the dataset distribution, ensuring that rare clause categories are not ignored during optimization.

Thus, the overall system combines:
- **Architectural innovation** (feature-aware gating)
- **Optimization-aware design** (weighted loss)

## 5. System-Level Design Thinking
The architecture is also designed with practical constraints in mind. Given the computational cost of transformer models, we limit sequence length and batch size to ensure efficient training without sacrificing contextual coverage.

This reflects a balance between model capacity and system constraints, making the approach scalable and deployable in real-world settings.

Unlike standard applications of pre-trained transformers, our model introduces a custom feature-aware fusion mechanism that integrates structural signals into the contextual embedding space. This represents a lightweight yet effective architectural innovation tailored to the legal NLP domain.
